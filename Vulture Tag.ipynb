```python
import pandas
import pickle

def load_obj(filename):
    with open(filename, 'rb') as handle:
        b = pickle.load(handle)
        return b

sample2 = load_obj("data_to_classify.pickle")
```

```python
def is_substantive(html_desc):
    if html_desc is None:
        return False
    char_count = len(html_desc)
    desc = html_desc.lower()
    if ((char_count<=5) & ("tp" not in desc) & ("hc" not in desc) & ("sc" not in desc & ("trade" not in desc)) :
        return False
    return True      
        
def sufficient_data(vol_data):
    desc = vol_data["description"]
    deck = vol_data["deck"]
    issue_count = vol_data["count_of_issues"]
    if is_substantive(deck) or is_substantive(desc) or issue_count>1:
        return True
    return False

def marvel(vol_data):
    if vol_data['publisher'] is not None:
        if vol_data['publisher']['id'] == 31:
            return True
        else:
            return False
    else:
        return False
```
