```python
import urllib2
import json
resource_hash = {"volume":"4050","issue":"4000","character":"4005"}
url_base = "http://comicvine.gamespot.com/api/"
key = "aff8790cd32512f45b429bb78cc21a7a87cf4d48"

def construct_url(i,resource,fields=""):
    new_base = url_base+resource+"/"+resource_hash[resource]+"-"+str(i)+"/"
    target = new_base + "?api_key="+key+"&format=json"
    if fields!="":
        target = target + "&field_list="+fields
    return target

def url_to_data( url ):
    req = urllib2.Request(url)
    req.add_header('User-agent', 'Mozilla 5.10')
    res = urllib2.urlopen(req)
    data = json.load(res)['results']
    return data

def look_up(i,resource,fields=""):
    return url_to_data(construct_url(i,resource,fields))

def vols_from_vill_id(vi,fields=""):
    vols = look_up(vi, "character",fields)["volume_credits"]
    return vols

def iss_from_vill_id(vi,fields=""):
    iss = look_up(vi, "character",fields)["issue_credits"]
    return iss    

def vols_to_add(character_list):
    target = []
    for character in character_list:
        char_id = character[1]
        vols = vols_from_vill_id(char_id)
        target = target+vols
    return target

#new_vols = vols_to_add(villains_to_add) 
```

```python
import pickle

def write_obj(obj,filename):
    with open(filename, 'wb') as handle:
        pickle.dump(obj, handle, protocol=pickle.HIGHEST_PROTOCOL)

def load_obj(filename):
    with open(filename, 'rb') as handle:
        b = pickle.load(handle)
        return b
```
