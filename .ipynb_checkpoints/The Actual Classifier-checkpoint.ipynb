```python
from textblob.classifiers import NaiveBayesClassifier
import numpy
import json
import nltk

def load_file(filename,t):
    f = open(filename,'rb')
    lines = f.read().splitlines()
    target = []
    if t == "int":
        for x in lines:
            target.append(int(x))
    else:
        if t == "json":
            for x in lines:
                y = json.loads(x)
                target.append(y)
        else:
            for x in lines:
                target.append(x)
    return target

i_black = load_file("i_black",'int')
i_white = load_file("i_white",'int')
v_black = load_file("v_black",'int')
v_white = load_file("v_white",'int')
collections = load_file("collections",'json')
originals = load_file("originals",'json')
```

```python
def test_train_data(source,label):
    train=[]
    t1=[]
    t2=[]
    i=0
    data_size = len(source)
    random_arr = numpy.random.choice(range(0,data_size),data_size)
    train_size = round(data_size*.64,1)
    devt_size = round(data_size*.16,1)
    target = []
    j = 0
    for i in random_arr:
        if j<train_size:
            target=train
        else:
            if j<train_size+devt_size:
                target=t1
            else:
                target=t2
        pair = (source[i],label)
        target.append(pair)
        j=j+1
    return (train,t1,t2)

c_tt = test_train_data(collections,"col")
o_tt = test_train_data(originals,"orig")

train = c_tt[0]+o_tt[0]

test1 = c_tt[1]+o_tt[1]
test2 = c_tt[2]+o_tt[2]
```

```python
def tag(char):
    if char == "<":
        return True
    else: return False

def advance(j,text):
    while True:
        letter=text[j]
        if letter == ">":
            return j
        j=j+1

def first_word(text):
    #target = json.dumps(text)
    #print target.split()[0]    
    target = ""
    j = 0
    if text is not None:
        while j < len(text):
            letter = text[j]
            if ((len(target)!=0) & (letter == " ")) or (len(target)==len(text)):
                #print target
                return unicode(target)
            if tag(letter):
                j=advance(j,text)
            else:
                target=target+letter
            j = j+1    
                    

def volume_features(vol_data):
    #words_in_title = words(vol_data['name'])
    #words_in_desc = words(vol_data['name'])
    features= {'issue_count': vol_data['count_of_issues'],
            'start_year': vol_data['start_year'],
            'high_issue_count': vol_data['count_of_issues']>=30,
            'first_word': first_word(vol_data['description']),
            'name_contains(by)': 'by' in vol_data['name'].lower(),
            'desc_contains(collected)': 'collected' in unicode(vol_data['description']).lower()
           }
    return features
```

```python
from HTMLParser import HTMLParser
import string
from textblob.utils import strip_punc

class MLStripper(HTMLParser):
    def __init__(self):
        self.reset()
        self.fed = []
    def handle_data(self, d):
        self.fed.append(d)
    def get_data(self):
        return ''.join(self.fed)

def strip_tags(html):
    s = MLStripper()
    s.feed(html)
    return s.get_data()

def corpus_from_text(data,key):
    #print(data)
    ls = []
    if data[key] is not None:        
        desc = strip_tags(data[key]).lower()
        desc = desc.split()
        for word in desc:
            word = strip_non_ascii(word)
            word = strip_punc(word,all=True)
            #print(word)
            if word not in ls:
                ls.append(word)
    return ls

                
def strip_non_ascii(string):
    stripped = (c for c in string if (0 < ord(c) < 127))
    return ''.join(stripped)

def create_corpus(cols,orls,key):
    ls = []
    for vol in cols:
        ls=ls+corpus_from_text(vol,key)
    for vol in orls:
        ls=ls+corpus_from_text(vol,key)
    return ls

c = []
print(corpus_from_text(collections[0],"description"))
title_corpus = create_corpus(collections,originals,"name")
desc_corpus = create_corpus(collections,originals,"description")
print title_corpus
```
