```python
def filtered_issues_to_csv(data,villain):
    file_name = villain+"_issues.txt"
    f = open(file_name,'w')
    f.write('volume_name;date;id;villain;url\n')
    for issue in data:
        name=issue['volume']['name']
        date=issue['cover_date']
        i=issue['id']
        row= name+';'+str(date)+';'+str(i)+';'+issue['villain_name']+';'+issue['site_detail_url']
        f.write("%s\n" % row)
```

```python
import time

def char_url_to_issues(url):
    url = url+"?api_key="+key+"&format=json&field_list=name,id,issue_credits,character_enemies"
    print(url)
    data = url_to_data(url)
#    print(data['results'])
    return(data)


def data_to_issues(data):
#    vill = octopus
#    url = vill['api_detail_url
#    data = char_url_to_issues(str(url))
    iss_links = []
    for issue in data['results']['issue_credits']:
        iss_links.append(issue['api_detail_url'])
    return(iss_links)
        
def iss_to_data(api_url):
    url = api_url+"?api_key="+key+"&format=json&field_list=name,id,cover_date,volume,site_detail_url"
    data=url_to_data(url)
#    print(data['results'])
    issues = json_to_array(data, False, 'issue', ['volume','cover_date','id','site_detail_url'])
    return(issues)
#    print(issues)

def add_to_dicts_in_array(data,title,name):
    for i in data:
        i[title]=name
    return data
```
